## Importing Libraries

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

# Data Visualization

## > Line Plot

* Line plot, also known as a line graph, is a type of data visualization that displays data points connected by straight line segments. It is commonly used to show the relationship between two continuous variables, such as time and temperature, or time and stock prices.

```python
sns.lineplot(x='timepoint',y='signal', data=fmri)
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal', data=fmri, hue='event')
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal', data=fmri, hue='event', style='event')
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal', data=fmri, hue='event', style='event', markers=True)
plt.show()
```
